Operators in Python

```python
x = 10; y =3
sum = x + y
subtraction = x - y
multiplication = x * y
division = round(x / y, 2) # round the result to 2 decimal places
remainder = x % y
print(sum, subtraction, multiplication, division, remainder)
```

Comparison

```python
x = 10; y =3

greater_than = x > y
greater_than_or_equal = x >= y
less_than = x < y
less_than_or_equal = x <= y
equal_to = x == y
not_equal_to = x != y
print(greater_than, greater_than_or_equal)
print(less_than, less_than_or_equal)    
print(equal_to, not_equal_to)

```

Logical operator

```python
x = 10 ; y = 3 ; z = 5;
result = (x > y) and (y < z) # True and True -> True
print(result)
```

```python
x = 10 ; y = 3 ; z = 5;
result = (x > y) or (y < z)
print(result)
```
